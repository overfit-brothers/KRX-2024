```python
import re

def clean_text(text):
    # 1. 이미지 태그 삭제
    text = text.replace('(image)\n\n(/image)','')
    text = text.replace('\u2002','')
    text = text.replace('\u2003','')
    # 2. 연속된 개행문자 하나로 변경 
    text = re.sub(r'\n\s*\n', '\n', text)
    
    # 3. 페이지 번호 삭제 (예: "001 증권시장의 이해" 형태)
    text = re.sub(r'\d{3}\s+[^\n]+$', '', text, flags=re.MULTILINE)
    
    # 4. '|'로 구분된 소제목 추출
    sections = []
    if '│' in text:
        parts = text.split('│')
        for part in parts[1:]:  # 첫 번째 부분 제외
            sections.append(part.strip())
            
    return {
        'cleaned_text': text.strip(),
        'sections': sections
    }
```

```python
all_text = ''
for data in raw:
    text = data['text']
    all_text += text
cleaned = clean_text(all_text)
```

```python
def chat(example):
    chat = PROMPT+'\n'+example
    response = client.beta.chat.completions.parse(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "You are a helpful assistant who is good at extracting information from text with json format."},
            {"role": "user", "content": chat},
        ],
        response_format=ConceptList,
    )
    answer_str = response.choices[0].message.content
    
    return answer_str
```

```python
import json
res_list = []

for data in cleaned['sections']:
    res_text = chat(data)
    res_json = json.loads(res_text)
    res_list.extend(res_json["concepts"])
```
